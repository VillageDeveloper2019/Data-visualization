$1. STory:$ The story here is to compare/rank legal drinking age in different parts of Canada.

$2. Problem:$ Bar chart is a good way to show this comparison but what we see is problem with effective bar chart. For one, the x-axis labels are not legible and are also superimposed in the bars. The bars are not sorted before plotting which makes the order of the bars disorganized. Another important thing to take note is that, this bar chart is not intended for ranking the deviation in drinking age so not starting the y-axis from zero could illude human preattentive processing.

$3. Alternative:$ We could improve on this chart by:

    a. starting the y-axis from zero
    
    b. removing the x-axis labels from the bars
    
    c. making the x-axis legible to read
    
    d. sorting the values before plotting the chart.
    

```python
import pandas as pd

age = [19, 18, 19, 18, 19, 18, 19, 19, 19, 19, 19, 19, 19]
width = 1/1.5
xs = list(range(len(age)))

area = ["B.C", "Alberta", "Saskatchewan", "Manitoba", "Ontario", "Quebec", "New Brunswick", "Pei", 
           "Nova Scotia", "Newfoundland", "NWT", "Nunavut", "Yukon"]

#put the lists in a dataframe and sort by age
df = pd.DataFrame(list(zip(area, age)), columns =['Area', 'Age']) 
df.sort_values(by = 'Age', inplace = True)               

figure, axes = plt.subplots(figsize = (20, 8))

plt.bar(df.Area, df.Age, width, color="dimgray", align="center")

plt.title('Drinking age across Canada', fontsize=24)
plt.xlabel('Provinces and Territories', fontsize=18)
plt.xticks(xs)
plt.xticks(rotation=30, fontsize = 14)
plt.ylabel('Age(years)', fontsize=18)
plt.ylim(0, 20)
plt.yticks(fontsize = 14)
#plt.grid(b=True)
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='black')
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')

plt.show()
plt.close()
```
